```python
import pandas as pd
```

```python
df_feitos = pd.read_excel("Feitos.xlsx")

df_prob = pd.read_excel("df_giov.xlsx")
print(len(df_prob))

df_prob2 = pd.read_excel("df_giov2.xlsx")
print(len(df_prob2))

df_disp = pd.read_excel('df_dispensados.xlsx')
print(len(df_disp))

df_prob = pd.concat([df_prob, df_prob2, df_disp]).reset_index(drop=True)
df_prob = df_prob.drop_duplicates(subset=['chat','dia'])
df_prob = df_prob.reset_index(drop=True)
df_prob = df_prob.sort_values(['nome', 'dia'])
print(len(df_prob))

display(df_prob.head())
```
