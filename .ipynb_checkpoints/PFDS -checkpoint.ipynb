# ANTIBIOTICO:

```python
from datetime import datetime, timedelta

def diferenca_horas(horario_inicial, horario_final):
    formato = "%H:%M"
    horario_inicial = datetime.strptime(horario_inicial, formato)
    horario_final = datetime.strptime(horario_final, formato)
    diferenca = horario_final - horario_inicial
    if diferenca.days < 0:
        diferenca = horario_inicial - horario_final
        return diferenca.total_seconds() / 60 * (-1)
    else:
        return diferenca.total_seconds() / 60
```

```python
for linha in range(len(df)):
    if df['controle'][linha] == 'prob' or pd.isnull(df['Antibiótico'][linha]):
        continue
    print('*'*50)

    print(df['atendimento'][linha])
    
    # Horario do inicio da cirurgia
    texto = df['texto_resto'][linha]
    horario_cirurgia = re.findall(r'Início cirurgia \d{2}/\d{2}/\d{4} \d{2}:\d{2}', texto)[0]
    horario_cirurgia = horario_cirurgia.replace('Início cirurgia ','')[11:]
    print('H_C:', horario_cirurgia)
    
    ######################################################################################
    
    # Horario do antibiotico
    dicionario = df['Antibiótico'][linha]
    palavra_grande_dms = 'Antibiótico realizado na sala operatória'
    if 'Tipo' in dicionario.keys() and palavra_grande_dms in dicionario.keys(): 
        if dicionario['Tipo'] == 'Profilático':
            frase = dicionario[palavra_grande_dms].upper().replace("H", ":")
            texto = re.findall(r'\d{2}:\d{2}', frase)
            if len(texto) > 0:
                horario_atb = texto[0]
            else:
                texto = re.findall(r'\d{2}/\d{2}/\d{4} \d{2}:\d{2}', frase)
                if len(texto) > 0:
                    horario_atb = texto[0][11:]
                else:
                    print("sem horario escrito")
                    continue
    else:
        print("sem itens localizados")
        continue

    ######################################################################################
    
    if horario_atb:  # Verifica se horario_atb foi definido
        print('H_ATB:', horario_atb)
        
        diferenca_tempo = diferenca_horas(horario_atb, horario_cirurgia)
        print(diferenca_tempo)
        
        if diferenca_tempo > 60 or diferenca_tempo < 0:
            print("Falha")
    else:
        print("Horário do antibiótico não encontrado.")
```

# GLICEMIA:

```python
def remover_horas(texto):
    texto_modificado = re.sub(r'\b\d{2}:\d{2}\b', '', texto)
    return texto_modificado

def contem_numero(texto):
    if re.search(r'\b\d{2,3}[^\d]*\b', texto):
        return True
    else:
        return False
```

```python
for linha in range(len(df)):
    if df['controle'][linha] == 'prob' or pd.isnull(df['Antibiótico'][linha]):
        continue

    atendimento = df['atendimento'][linha]
    
    dicionario = df['Antibiótico'][linha]
    if 'Protocolo de glicemia no intraoperatório' in dicionario.keys():
        if dicionario['Protocolo de glicemia no intraoperatório'] == 'Sim':
            if 'Histórico de Glicemias' in dicionario.keys():
                print('"', dicionario['Histórico de Glicemias'], '"')
                ################-TRATAMENTO FINAL-#################
                texto = dicionario['Histórico de Glicemias']
                texto = remover_horas(texto.upper().replace("H", ":"))
                if contem_numero(texto):
                    print(atendimento, '->', 'ok - fez tudo certo')
                else:
                    print(atendimento, '->', 'falha'*15)
                print('\n')
            else:
                print(atendimento, '->', 'falha'*15)
                break
        else:
            continue #print(atendimento, '->', 'ok - não abriu protocolo')
    else:
        continue #print(atendimento, '->', 'ok - não esta no dicionario')
```
