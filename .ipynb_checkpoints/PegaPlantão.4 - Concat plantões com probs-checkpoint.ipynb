```python
import pandas as pd
import numpy as np

import warnings
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

import unidecode

import re

from datetime import date #para obter a data-dia atual
```

```python
df = pd.read_excel('tabela_erros.xlsx')

print(len(df))
```

```python
df
```

```python
#data do plantão
data = []
for i in range(0, len(df['chat'])):
    loc = df['chat'][i].find(']')+1
    loc2 = df['chat'][i][loc+1:].find(':')+1
    if df['chat'][i][loc+loc2+2+2] == "/" or df['chat'][i][loc+loc2+2+1] == "/":
        data.append(re.sub(r'[^0-9:,/]', '', df['chat'][i][loc+loc2+2:loc+loc2+7]))
    else:
        data.append("ind")
df['data'] = data
df['data'].unique()
```

```python
#nome do profissional
nome = []
count = 0
for i in range(0, len(df['chat'])):
    loc = df['chat'][i].find(']')+1
    loc2 = df['chat'][i][loc+1:].find(':')+1
    nome.append(unidecode.unidecode(df['chat'][i][loc+1:loc+loc2].upper()))
    count += 1
    #print(count, "->", unidecode.unidecode(chat[i][loc+1:loc+loc2].upper()))
df['nome'] = nome
df['nome'].unique()
```

```python
#mensagem isolada
mensagem = []
for i in range(0, len(df['chat'])):
    loc = df['chat'][i].find(']')+1
    loc2 = df['chat'][i][loc+1:].find(':')+1
    mensagem.append(df['chat'][i][loc+loc2+2:].strip())
df['mensagem'] = mensagem
df['mensagem'].head()
```

```python
for nome in df["nome"].unique():
    print(nome, "="*10)
    selecao = df.loc[df['nome']==nome]
    selecao.reset_index(drop=True, inplace=True)
    
    for i in range(len(selecao)):
        print(selecao["mensagem"][i])
    
    print("\n")
```

# Organizando

```python
df = pd.read_excel("tabela_erros_todos_01a"+str(date.today().day)+".xlsx")
df
```

```python
#mensagem isolada
mensagem = []
for i in range(0, len(df['chat'])):
    loc = df['chat'][i].find(']')+1
    loc2 = df['chat'][i][loc+1:].find(':')+1
    mensagem.append(df['chat'][i][loc+loc2+2:].strip())
df['mensagem'] = mensagem
df['mensagem'].head()

df.head()
```

```python
print("Quantidade de Registros:", len(df))
print("Quantidade de Registros(pessoas) :", len(df['nome'].unique()))
```

```python
for nome in sorted(list(df['nome'].unique())):
    print(nome, "="*10)
    selecao = df.loc[df['nome']==nome]
    selecao = selecao.sort_values(by='mensagem')
    selecao.reset_index(drop=True, inplace=True)
    
    for i in range(len(selecao)):
        print(selecao["mensagem"][i])
    
    print("\n")
```
