```python
import pandas as pd
import numpy as np

import warnings
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

import unidecode

import re
```

```python
def read_file(file):
    '''Reads Whatsapp text file into a list of strings'''
    x = open(file,'r', encoding = 'utf-8')
    y = x.read()
    content = y.splitlines()
    return content

chat = read_file('_chat.txt')
chat
```

# CAPTURANDO DADOS 1:

```python
nome = []
count = 0
for i in range(0, len(chat)):
    loc = chat[i].find(']')+1
    loc2 = chat[i][loc+1:].find(':')+1
    nome.append(unidecode.unidecode(chat[i][loc+1:loc+loc2].upper()))
    count += 1
    #print(count, "->", unidecode.unidecode(chat[i][loc+1:loc+loc2].upper()))
set(nome)
```

```python
dia = []
for i in range(0, len(chat)):
    dia.append(chat[i][1:11])
set(dia)
```

```python
data = []
for i in range(0, len(chat)):
    loc = chat[i].find(']')+1
    loc2 = chat[i][loc+1:].find(':')+1
    if chat[i][loc+loc2+2+2] == "/" or chat[i][loc+loc2+2+1] == "/":
        data.append(re.sub(r'[^0-9:,/]', '', chat[i][loc+loc2+2:loc+loc2+7]))
    else:
        data.append("ind")
set(data)
```

```python
mensagem = []
for i in range(0, len(chat)):
    loc = chat[i].find(']')+1
    loc2 = chat[i][loc+1:].find(':')+1
    if data[i] == "ind":
        mensagem.append(unidecode.unidecode(chat[i][loc+loc2+2:].upper().strip()))
    else:
        mensagem.append(unidecode.unidecode(chat[i][loc+loc2+7:].upper().strip()))
set(mensagem)
```

```python
df = pd.DataFrame({'mensagem':chat,
                   'datamensagem':dia,
                   'nome':nome,
                   'data':data,
                  'filtro':mensagem})

df['filtro'] = df['filtro'].str.upper()
df[30:37]
```

# CAPTURANDO DADOS 2:

```python
df['filtro2'] = df['filtro']
```

```python
#ARRUMANDO NOMES PARA EVITAR PROBLEMAS
df['filtro2'] = df['filtro2'].apply(lambda x: x.replace(' SLO ', ' OSASCO '))
df['filtro2'] = df['filtro2'].apply(lambda x: x.replace(' SLO', ' OSASCO'))
df['filtro2'] = df['filtro2'].apply(lambda x: x.replace('SLO ', 'OSASCO '))
df['filtro2'] = df['filtro2'].apply(lambda x: x.replace('(SLO', ' OSASCO'))
df['filtro2'] = df['filtro2'].apply(lambda x: x.replace('SINO', 'OSASCO'))

#Definindo nome das unidades
unidade = ['ALPHA','ANALIA','AVICCENA','ITAIM','JABA','MATERNIDADE','MORUMBI','OSASCO','VILLA','VNS']

#Definindo nome das unidades
for i in unidade:
    df[str(i)] = [1 if str(i) in x else 0 for x in df['filtro2']]

df[30:37]
```
