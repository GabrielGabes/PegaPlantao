```python
import pandas as pd
import numpy as np

import warnings
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

import unidecode

import re
```

```python
def read_file(file):
    '''Reads Whatsapp text file into a list of strings'''
    x = open(file,'r', encoding = 'utf-8')
    y = x.read()
    content = y.splitlines()
    return content

chat = read_file('_chat.txt')
df = pd.DataFrame({'chat':chat})
df
```

# CAPTURANDO DADOS 1 =========================

```python
#dia da mensagem
dia = []
for i in range(0, len(df['chat'])):
    dia.append(df['chat'][i][1:11])
df['dia'] = dia
df['dia'].unique()
```

```python
#nome do profissional
nome = []
count = 0
for i in range(0, len(df['chat'])):
    loc = df['chat'][i].find(']')+1
    loc2 = df['chat'][i][loc+1:].find(':')+1
    nome.append(unidecode.unidecode(df['chat'][i][loc+1:loc+loc2].upper()))
    count += 1
    #print(count, "->", unidecode.unidecode(chat[i][loc+1:loc+loc2].upper()))
df['nome'] = nome
df['nome'].unique()
```

```python
#data do plantão
data = []
for i in range(0, len(df['chat'])):
    loc = df['chat'][i].find(']')+1
    loc2 = df['chat'][i][loc+1:].find(':')+1
    if df['chat'][i][loc+loc2+2+2] == "/" or df['chat'][i][loc+loc2+2+1] == "/":
        data.append(re.sub(r'[^0-9:,/]', '', df['chat'][i][loc+loc2+2:loc+loc2+7]))
    else:
        data.append("ind")
df['data'] = data
df['data'].unique()
```

```python
#mensagem isolada
mensagem = []
for i in range(0, len(df['chat'])):
    loc = df['chat'][i].find(']')+1
    loc2 = df['chat'][i][loc+1:].find(':')+1
    if data[i] == "ind":
        mensagem.append(unidecode.unidecode(df['chat'][i][loc+loc2+2:].upper().strip()))
    else:
        mensagem.append(unidecode.unidecode(df['chat'][i][loc+loc2+7:].upper().strip()))
df['mensagem'] = mensagem
df['mensagem'].head()
```

```python
df['filtro2'] = df['filtro']
df['filtro2'] = df['filtro2'].apply(lambda x: x.replace("H" , ':'))
df['filtro2'].unique()
```

```python
lista = []

#retirando todos caracteres exceto 0-9, : e H
for i in range(0, len(df)):
    string = df['filtro2'][i]
    lista.append(re.sub(r'[^0-9:,H]', '', string).strip())
    
df['filtro2'] = lista
df['filtro2'].value_counts()
```

```python
#separando apenas os formatos HH:MM
lista = []
for i in range(0, len(df)):
    string = df['filtro2'][i]
    lista.append(re.findall(r'[0-9][0-9]:[0-9][0-9]', string))
    
df['filtro3'] = lista
df['filtro3'].head()
```

# arrumação final

```python
#capturando index dos valores não reconhecidos
index_q = []
count = 0 
for i in range(0, len(df)):
    if len(df['filtro3'][i]) == 0:
        index_q.append(count)
        print(count, "->", df['filtro'][i])
    count+=1
```

```python
#limpando adicionando 0 no primeiro algorismo dos valores não reconhecidos
for i in index_q:
    if len(df['filtro3'][i]) == 0:
        if bool(re.findall(r'[0-9]:[0-9][0-9]', df['filtro2'][i])) == True:
            df['filtro3'][i] = "0"+re.findall(r'[0-9]:[0-9][0-9]', df['filtro2'][i])[0]
            print("True", i, '->', df['filtro2'][i], ' - correto:', df['filtro3'][i])
        else:
            print("False", i, '->', df['filtro2'][i], '-', df['filtro'][i]) #imprimindo ondem não deu para corrigir
```

```python
#tirando os valores unicos de uma lista
for i in range(0, len(df)):
    if type(df['filtro3'][i]) == list:
        if len(df['filtro3'][i]) == 1:
            df['filtro3'][i] = df['filtro3'][i][0]
```

```python
from datetime import date
from datetime import time
from datetime import datetime
from datetime import timedelta
```

```python
df_ok['filtro4'] = df_ok['filtro3']
df_ok['filtro5'] = df_ok['filtro3']
```

```python
df_ok['filtro4'] = df_ok['filtro4'].apply(lambda x: datetime.strptime(x, '%H:%M'))

print(df_ok['filtro4'].dt.year.unique())
print(df_ok['filtro4'].dt.month.unique())
print(df_ok['filtro4'].dt.day.unique())
print(df_ok['filtro4'].dt.hour.unique())
```

```python
#ARREDONDANDO OS HORARIOS
#entrada: arredondar para baixo
#saida: arrendodar para cima
df_ok['ERRO'] = ''

for i in range(0, len(df_ok)):
    print(i,'=======================')
    valor = int(df_ok['filtro3'][i][3:])
    print(df_ok['filtro3'][i])
    if valor != 0 and valor != 30:
        if valor < 30:
            if df_ok['tipo'][i] == "entrada": 
                print('entrada', valor, '->', 0)
                print('horario ajeitado ->', str(df_ok['filtro4'][i] + timedelta(minutes=-valor))[11:-3])
                df_ok['filtro5'][i] = str(df_ok['filtro4'][i] + timedelta(minutes=-valor))[11:-3]
            elif df_ok['tipo'][i] == "saida":
                print('saida', valor, '->', 30)
                print('horario ajeitado ->', str(df_ok['filtro4'][i] + timedelta(minutes=+(30-valor)))[11:-3])
                df_ok['filtro5'][i] = str(df_ok['filtro4'][i] + timedelta(minutes=+(30-valor)))[11:-3]
            else:
                df_ok['ERRO'][i] = 'ERRO'
                print('====================ERRO1====================')
        elif valor > 30:
            if df_ok['tipo'][i] == "entrada":
                print('entrada', valor, '->', 30)
                print('horario ajeitado ->', str(df_ok['filtro4'][i] + timedelta(minutes=-(valor-30)))[11:-3])
                df_ok['filtro5'][i] = str(df_ok['filtro4'][i] + timedelta(minutes=-(valor-30)))[11:-3]
                
            elif df_ok['tipo'][i] == "saida": 
                print('saida', valor, '->', 60)
                print('horario ajeitado ->', str(df_ok['filtro4'][i] + timedelta(minutes=+(60-valor)))[11:-3])
                df_ok['filtro5'][i] = str(df_ok['filtro4'][i] + timedelta(minutes=+(60-valor)))[11:-3]
            else:
                df_ok['ERRO'][i] = 'ERRO'
                print('====================ERRO2====================')
        else:
                df_ok['ERRO'][i] = 'ERRO'
                print('====================ERRO3====================')
    elif valor == 0 or valor == 30:
        df_ok['filtro5'][i] = df_ok['filtro3'][i]
    else:
        df_ok['ERRO'][i] = 'ERRO'
        print('====================ERRO4====================')
    print(df_ok['filtro5'][i])
    print('\n')
```

```python
#DETERMINANDO O HORARIO DE SAIDA E ENTRADA
df_ok['HORA1'] = ''
df_ok['HORA2'] = ''

for i in range(0, len(df_ok)):
    print(i,'=======================')
    if df_ok['tipo'][i] == 'entrada':
        df_ok['HORA1'][i] = df_ok['filtro5'][i]
        valor = int(df_ok['HORA1'][i][:2])
        if valor < 7:
            df_ok['HORA2'][i] = "07:00"
        elif valor >= 7 and valor < 13:
            df_ok['HORA2'][i] = "13:00"
        elif valor >= 13 and valor < 19:
            df_ok['HORA2'][i] = "19:00"
        else:
            df_ok['ERRO'][i] = "ERRO"
            print('====================ERRRO1====================')
            
    elif df_ok['tipo'][i] == 'saida':
        df_ok['HORA2'][i] = df_ok['filtro5'][i]
        valor = int(df_ok['HORA2'][i][:2])
        if valor >= 7 and valor < 13:
            df_ok['HORA1'][i] = "07:00"
        elif valor >= 13 and valor < 19:
            df_ok['HORA1'][i] = "13:00"
        elif valor >= 19:
            df_ok['HORA1'][i] = "19:00"
        elif valor >= 0 and valor < 7:
            df_ok['ERRO'][i] = "VERIFICAR"
        else:
            df_ok['ERRO'][i] = "ERRO"
            print('====================ERRRO2====================')
    else:
        df_ok['ERRO'][i] = "ERRO"
        df_ok['ERRO'][i] = "ERRO"
        print('====================ERRRO3====================')
        
    print('tipo:', df_ok['tipo'][i])
    print('horario dado =====>', df_ok['filtro3'][i], '->', df_ok['filtro5'][i])
    print('horario entrada:', df_ok['HORA1'][i])
    print('horario saida:', df_ok['HORA2'][i])
    print('\n')
```
