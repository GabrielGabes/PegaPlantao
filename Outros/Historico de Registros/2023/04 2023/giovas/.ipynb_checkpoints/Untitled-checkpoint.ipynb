```python
import os

# caminho para a pasta raiz
root_path = "C:/Users/Gabriel/Documents/giovas"

quantidade_giovanelli = 0
quantidade_gabriel = 0

# percorre a pasta raiz e todas as suas subpastas
for root, dirs, files in os.walk(root_path):
    # imprime o nome de todos os arquivos
    for file in files:
        if file.endswith(".xlsx"):
            endereco = str(os.path.join(root, file))
            if (endereco[-10:] == 'giov2.xlsx') or (endereco[-10:] == '_giov.xlsx'):
                df = pd.read_excel(endereco)
                quantidade_giovanelli += len(df)
            else:
                df = pd.read_excel(endereco)
                quantidade_gabriel += len(df)


print('quantidade_giovanelli:',quantidade_giovanelli)
print('quantidade_gabriel:',quantidade_gabriel)
```
